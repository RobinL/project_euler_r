The sequence of triangle numbers is generated by adding the natural numbers. So the 7th triangle number would be 1 + 2 + 3 + 4 + 5 + 6 + 7 = 28. The first ten terms would be:

1, 3, 6, 10, 15, 21, 28, 36, 45, 55, â€¦

Let us list the factors of the first seven triangle numbers:

1: 1

3: 1,3

6: 1,2,3,6

10: 1,2,5,10

15: 1,3,5,15

21: 1,3,7,21

28: 1,2,4,7,14,28
We can see that 28 is the first triangle number to have over five divisors.
What is the value of the first triangle number to have over five hundred divisors?

```r
get_primes <- function(max_prime){
   
    possible_primes <- seq(3,max_prime,2) #Odd primes 
    i <- 1
    repeat {
        #If the next number to seive is more than half of the maximum prime, then we're done
        if (possible_primes[i]>max_prime**0.5) break  

        #Otherwise filter out all multiples of this number from this list 
        possible_primes <- possible_primes[(possible_primes %% possible_primes[i] != 0) | possible_primes==possible_primes[i]]
        i <- i+1
    }
    c(2,possible_primes)
}
```

```r
primes <- get_primes(500000)

get_prime_factors <- function(num) {
    
    primes <- primes[primes<(ceiling(num^0.5))]
    factors <- integer()
    
    for (p in primes) {
        
        if (p>num) break
            
         if (num %% p ==0) {
            num <- num/p
            factors <- c(factors, p)
        } 
    }
    factors
}
```

```r
get_factors <- function(num){

    p_factors <- get_prime_factors(num)
    factors <- p_factors
    
    recurse <- function(num) {
        
        for (p in p_factors) {
            if (num %% p == 0 & !((num/p) %in% factors)) {
                factors <<- c(factors,num/p)
                recurse(num/p)
            }
        }   
    }
    
    recurse(num)
    sort(c(1,factors))
}
```

```r
num_factors <-function(num){
    length(get_factors(num))
}


found <- FALSE

for (i in 3:100000) {
    tri_num <- (i*(i+1))/2
    if (num_factors(tri_num)>500) {
        break
    }
}
tri_num
```
