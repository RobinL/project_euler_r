The following iterative sequence is defined for the set of positive integers:

n → n/2 (n is even)
n → 3n + 1 (n is odd)

Using the rule above and starting with 13, we generate the following sequence:

13 → 40 → 20 → 10 → 5 → 16 → 8 → 4 → 2 → 1
It can be seen that this sequence (starting at 13 and finishing at 1) contains 10 terms. Although it has not been proved yet (Collatz Problem), it is thought that all starting numbers finish at 1.

Which starting number, under one million, produces the longest chain?

NOTE: Once the chain starts the terms are allowed to go above one million.

```r
get_chain_size_closure <- function() {
    
    #To store pre-computed chain lengths
    cache <- numeric()
    
    next_term <-function(n) {
        if (n%%2==0) return(n/2)
        return(3*n+1)
    }
        
    get_chain_size <- function(start_term,debug=FALSE) {
        
        if (debug)  print(cache)
        this_chain<-start_term
        terms_remaining_from_this_term <-0
        #Repeat until we get to 1, or find a cached term
        term <- start_term
        repeat {
            
            if (!(is.na(cache[term]))) {
               terms_remaining_from_this_term <- cache[term]
               break
            }
            
            term <- next_term(term)
            this_chain <-c(this_chain, term)
            
            if (term==1) break
            
            
        }
            
       #Write all new terms to cache
       len <- length(this_chain)
            
       for (i in 1:len) {
           t<-this_chain[i]
           if (t<1000000) {
               cache[t] <<- (len-i) + terms_remaining_from_this_term  
           }   
       }  
           
            
        #Return length
        len + terms_remaining_from_this_term
        
        
    }
    
    
    
}
get_chain_size <-get_chain_size_closure()
        
longest <-1
longest_index <- 1        
for (i in 1:1e6) {
    this_length <- get_chain_size(i)
    if (this_length > longest){
        longest <- this_length
        longest_index <- i 
    }
}
```
